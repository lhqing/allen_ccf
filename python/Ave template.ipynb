```python
import nrrd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('figure', dpi=200)
```

## Load nrrd

```python
ave_template_data, meta = nrrd.read('../ccf/average_template_25.nrrd')
```

```python
type(ave_template_data)
```

```python
# AP, SI, LR
ave_template_data.shape
```

## Slice

```python
# cornal slice
slice_data = ave_template_data[300, :, :]
print(slice_data.shape)
plt.imshow(slice_data)
```

```python
# Axial slice
slice_data = ave_template_data[:, 200, :]
print(slice_data.shape)
plt.imshow(slice_data)
```

```python
# sagittal slice
slice_data = ave_template_data[:, :, 300]
print(slice_data.shape)
plt.imshow(slice_data)
```

```python
plt.imshow(slice_data.T)
```

```python
plt.imshow(slice_data.T[:, ::-1])
```

```python
plt.imshow(slice_data.T[:, ::-1], vmin=0, vmax=200)
```

## Chunk

```python
chunked_data = ave_template_data[300:324, :, :]
print(chunked_data.shape)
```

```python
chunked_data = ave_template_data.copy()
chunked_data[:300, :, :] = 0
chunked_data[324:, :, :] = 0
print(chunked_data.shape)
```

```python
plt.imshow(chunked_data[:, :, 300].T, vmin=0, vmax=200)
```
